```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

def k_medoids(data, k, max_iterations=100):
    # Randomly initialize k medoids from the data points
    medoids_idx = np.random.choice(len(data), k, replace=False)
    medoids = data[medoids_idx]
    
    medoid_history = []
    distance_history = []
    cluster_history = []

    for _ in range(max_iterations):
        # Calculate distances from each data point to each medoid
        distances = np.array([euclidean_distance(data_point, medoid) for data_point in data for medoid in medoids])
        distances = distances.reshape(len(data), k)
        
        # Assign each data point to the nearest medoid
        labels = np.argmin(distances, axis=1)

        # Create clusters with medoids as centers
        new_medoids = np.array([data[labels == i].mean(axis=0) for i in range(k)])

        # Find the medoid within each cluster that minimizes the sum of distances to other points
        for i in range(k):
            cluster_points = data[labels == i]
            total_distance = np.sum([np.sum([euclidean_distance(medoid, point) for point in cluster_points]) for medoid in cluster_points])
            medoids[i] = cluster_points[np.argmin([np.sum([euclidean_distance(medoid, point) for point in cluster_points]) for medoid in cluster_points])]
        
        if np.all(medoids == new_medoids):
            break

        medoid_history.append(medoids)
        distance_history.append(distances)
        
        # Save clustering points in this iteration
        cluster_points = [data[labels == i] for i in range(k)]
        cluster_history.append(cluster_points)

    medoid_history.append(medoids)
    distance_history.append(distances)

    return labels, medoid_history, distance_history, cluster_history

# Example usage
X = [8, 24, 15, 4, 24]
Y = [4, 4, 8, 4, 12]
data = np.column_stack((X, Y))

k = 2
labels, medoid_history, distance_history, cluster_history = k_medoids(data, k)

# Print medoids and distances from medoids for each iteration
for i, medoids in enumerate(medoid_history):
    print(f"ITERATION {i + 1}:")
    print(f"\nMedoids:\n {medoids}")
    print("\nDistances from Medoids:")
    cluster_labels = labels + 1  # Adjust index to start from 1 instead of 0
    df = pd.DataFrame(distance_history[i], columns=[f"Medoid {j+1}" for j in range(k)])
    df.insert(0, "Data Point", data.tolist())
    df["Cluster"] = cluster_labels
    print(df.to_string(index=False))
    print("\n")

    # Print points in each cluster for this iteration
    if i < len(cluster_history):
        print("Cluster Points:")
        for j, cluster_points in enumerate(cluster_history[i]):
            print(f"Cluster {j + 1}:")
            print(cluster_points.tolist())
    print("\n")

# Print final clusters
print("Final Clusters:")
df = pd.DataFrame(data, columns=["X", "Y"])
df["Cluster"] = labels + 1  # Adjust index to start from 1 instead of 0
print(df.to_string(index=False, header=True))

# Plot clusters
plt.scatter(data[:, 0], data[:, 1], c=labels)
plt.scatter(medoid_history[-1][:, 0], medoid_history[-1][:, 1], marker='x', c='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('K-medoids Clustering')
plt.show()
```
